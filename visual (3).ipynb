```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
```

```python
corona = pd.read_excel('corona2.xlsx')
```

```python
corona.head(10)
```

```python
corona.info()
```

```python

sns.set(rc={'axes.facecolor':'black','figure.facecolor':'white'})
sns.set_context("poster",font_scale=0.58,rc={"grid.linewidth":1,"font.size":5,"axes.labelsize":15})

sns.lmplot(x="District", y="Total Cases", hue="District", data=corona,size=17).set(title="TOTAL CORONA CASES IN TAMILNADU - AUGUST (2021) REPORT")
```

```python
sns.set(rc={'axes.facecolor':'black','figure.facecolor':'white'})
sns.set_context("poster",font_scale=0.58,rc={"grid.linewidth":1,"font.size":15,"axes.labelsize":20})
sns.barplot(corona['District'], corona['Active Cases']).set(title="ACTIVE CASES IN TAMILNADU - AUGUST (2021) REPORT")
sns.set(rc={'figure.figsize':(28,13)})
```

```python
sns.set(rc={'axes.facecolor':'orange','figure.facecolor':'violet'})
sns.set_context("poster",font_scale=0.58,rc={"grid.linewidth":2,"font.size":15,"axes.labelsize":20})
sns.lineplot(corona['District'], corona['Total Deaths']).set(title="TOTAL DEATHS IN TAMILNADU - AUGUST (2021) REPORT")
sns.set(rc={'figure.figsize':(28,9)})
```

```python

sns.set(rc={'axes.facecolor':'orange','figure.facecolor':'violet'})
sns.set_context("poster",font_scale=0.58,rc={"grid.linewidth":2,"font.size":15,"axes.labelsize":20})
sns.violinplot(corona['District'], corona['Total Recovered']).set(title="TOTAL RECOVERIES IN TAMILNADU - AUGUST (2021) REPORT")
sns.set(rc={'figure.figsize':(28,10)})
```
